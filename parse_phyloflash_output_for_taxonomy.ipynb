### First extract only the taxon names ('name') and 'new_est_reads' columns from the Bracken raw output

```python
pf = pd.read_csv("1903c111_10m-1_bmtag.phyloFlash.extractedSSUclassifications.csv")
pf.head()
```

### Filter out eukaryotic assignments

```python
pf = pf[pf['taxonomy'].str.contains('Eukaryota')==False]
pf.head()
```

```python
taxa = pf[['taxonomy']].copy()
taxa[['Domain','Phylum','Class','Order','Family','Genus','Species']] = taxa['taxonomy'].str.split(";", n=7, 
                                                                                                expand=True)
taxa.head()
```

```python
for blast in glob.glob("*.csv"):
    a, b, c, d = blast.split('.')
    print(a, b, c, d)
```

```python
for pf in glob.glob("*.csv"):
    a, b, c, d = pf.split('.')
    pf = pd.read_csv(pf)
    pf = pf[pf['taxonomy'].str.contains('Eukaryota')==False]
    taxa = pf[['taxonomy']].copy()
    taxa[['Domain','Phylum','Class','Order','Family','Genus','Species']] = taxa['taxonomy'].str.split(";", n=7, expand=True)
    taxa.to_csv('%s-species.tsv' % a, index=None, sep='\t')
```
